```python
import pandas as pd
from pycaret.classification import *
```

```python
data = pd.read_csv('./hackathon_all.csv', sep=';')
data
```

```python
data.drop(labels=['Лекарство / давление', 'Лекарство / холестерин', 'Лекарство / инсульт', 'Лекарство / диабет', 'Лекарство / астма', 'Лекарство / Китай', 'Лекарство / прочие', 'Лекарство / прочие - уточнение', 'Лекарство / неизвестно', 'Перечень лекарства'], inplace = True, axis=1)

```

```python
data['target']= data['Target: Стенокардия, ИБС, инфаркт миокарда']
data.drop('Target: Стенокардия, ИБС, инфаркт миокарда', axis=1)
```

```python
df = pd.concat([data.iloc[:,1:-11], data.iloc[:,-1:]], axis=1)
df
```

```python
exp = setup(data = df, target = 'target', session_id=123, remove_perfect_collinearity=False, numeric_imputation='zero')
```

```python
best_model = compare_models()
```

```python
dt = create_model('dt')
```

```python
dt_tuned = tune_model(dt)
```

```python
evaluate_model(dt_tuned)
```
