```python
data['Duration: Длительность артериальной гипертензии'] = data['Длительность артериальной гипертензии']
data['Duration: Даность ОНМК'] = data['Даность ОНМК']
data['Duration: Длительность стенокардии, ИБС, инфаркта миокарда'] = data['Длительность стенокардии, ИБС, инфаркта миокарда']
data['Duration: Сердечная недостаточность'] = data['Сердечная недостаточность']
data['Duration: Длительность прочих заболеваний сердца'] = data['Длительность прочих заболеваний сердца']

data['Target: Артериальная гипертензия'] = data['Артериальная гипертензия']
data['Target: ОНМК'] = data['ОНМК']
data['Target: Стенокардия, ИБС, инфаркт миокарда'] = data['Стенокардия, ИБС, инфаркт миокарда']
data['Target: Сердечная недостаточность'] = data['Сердечная недостаточность']
data['Target: Прочие заболевания сердца'] = data['Прочие заболевания сердца']

cols = list(data.columns)
cols.remove('Артериальная гипертензия')
cols.remove('ОНМК')
cols.remove('Стенокардия, ИБС, инфаркт миокарда')
cols.remove('Сердечная недостаточность')
cols.remove('Прочие заболевания сердца')

cols.remove('Длительность артериальной гипертензии')
cols.remove('Даность ОНМК')
cols.remove('Длительность стенокардии, ИБС, инфаркта миокарда')
cols.remove('Длительность сердечной недостаточности')
cols.remove('Длительность прочих заболеваний сердца')



data = data[cols]
data
```
